```python
#Ask user to set perimeter
start = int(input('Put in a number you want to start with: '))
end = int(input('Put in a number you want to end with: '))
#Loop through the perimeter
for x in range(start,end+1):
    
    #if divisible by 15
    if x%15 == 0:
        print('FizzBuzz')
    
    #if divisible by 3
    elif x%3 == 0:
        print('Fizz')
    
    #if divisible by 5
    elif x%5 == 0:
        print('Buzz')
    
    #normal numbers
    else:
        print(x)
```
